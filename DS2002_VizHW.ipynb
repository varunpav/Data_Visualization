```python
import pandas as pd
import matplotlib.pyplot as plt

# Downloaded and connected csv this way
url = 'Top_Highest_Openings.csv'
df = pd.read_csv(url)

# Bar Chart for Opening Weekend Gross
# Taking only the top 20 movies by opening gross for clarity
top_movies = df.nlargest(20, 'Opening')
plt.figure(figsize=(15, 8))  # Increased figure size for readability
plt.bar(top_movies['Release'], top_movies['Opening'], color='blue')
plt.xlabel('Movie', fontsize=14)
plt.ylabel('Opening Weekend Gross ($)', fontsize=14)
plt.title('Top 20 Movies by Opening Weekend Gross', fontsize=18)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()

# Histogram for the Distribution of Total Gross
plt.figure(figsize=(10, 6))
plt.hist(df['Total Gross'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Total Gross ($)')
plt.ylabel('Number of Movies')
plt.title('Distribution of Total Gross of Movies')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Scatter Plot for Theaters vs. Opening Gross
plt.figure(figsize=(10, 6))
plt.scatter(df['Theaters'], df['Opening'], color='green')
plt.xlabel('Number of Theaters')
plt.ylabel('Opening Gross ($)')
plt.title('Relationship Between Number of Theaters and Opening Gross')
plt.grid(True)
plt.show()
```
